# Parte 3: funciones

```python

def funcion_que_encuentra_digitos(mi_string):
    digitos = []
    for i in range(len(mi_string)):
        if (mi_string[i]).isdigit():
            digitos.append(i)
    return digitos


digi = funcion_que_encuentra_digitos("Yo nací en 1986")


print("El texto tenía digitos en las posiciones: "+str(digi))
```

### Ejercicio 1: Procedimiento que imprime *n* términos de la serie de Fibonacci

```python
def fibonacci(n):
    a,b = 0,1
    print (a)
    print (b)
    while a < n:
        c = a + b
        a = b
        b = c
        print (c)
```

```python
fibonacci(4)
```

### Ejercicio 2: Función que calcula el factorial de *n*

```python
def factorial(n):
    a = 1
    for i in range(n):
        a = a * (i + 1)
    
    print (a)
```

```python
factorial(3)
```

### Ejercicio 3: Función que lee *n* y calcula Fibonacci (si par) o factorial (si impar)

```python
def lee_n(n):
    if n%2 == 0:
        fibonacci(n)
    else:
        factorial(n)
```

```python
lee_n(1)
```

### Ejercicio 5: Analytics sin Pandas (¿?)

Suponga que le pasan un archivo ´Pickle´ que contiene un diccionario con información sobre una variable de interés para una empresa.

Los "keys" del diccionario indican el mes y año asociado a un reporte, mientras que cada "value" contiene una lista con las observaciones de la variable en cada día de ese mes y año.

```python
import pickle as pkl

with open("Datos_2002_2009.pkl", "rb") as f:
    datos = pkl.load(f)

# ver "llaves"
```

```python
#ver ALGUNOS valores
list(datos.items())[0:1]
```

### Misiones:

1. Garantizar que los valores de las listas sean válidos y numéricos
2. Permitir acceder a los datos por año y mes (en lugar de un *string* con ambas)
3. Construir un nuevo diccionario con los promedios y desviaciones por mes, y un diagrama de barras de los promedios
4. Construir un nuevo diccionario con los promedios y desviaciones por año, y un diagrama de barras de los promedios

### Solución:

## 1. Garantizar que los valores de las listas sean válidos y numéricos

```python
# para cada dato de cada VALUE ver si es numerico (corregir)

# hint: isdigit

def my_datos_only_numbers(datos):
    for values in datos.values():
        for value in values:
            if type(value) != int:
                if str(value).isdigit():
                    values[values.index(value)] = int(value.strip())
                else:
                    values[values.index(value)] = 0
    return (datos)
    

# si no es numerico, tambien corregir
```

Comprobar si los datos quedaron bien procesados:

```python
# visualizar de nuevo ALGUNOS valores para corroborar
datos_validos = my_datos_only_numbers(datos)
print(datos_validos)
```

## 2. Permitir acceder a los datos por año y mes (en lugar de un *string* con ambas)

```python
# generar tuplas en lugar de keys planos (hint: split)

my_tuple = []
datos_new = {}
for key in datos:
    new_key = tuple(key.split('_'))
    datos_new[new_key] = datos[key]
    my_tuple.append(new_key)
```

```python
# set de años (sin repetir)
years = []
for i in my_tuple:
    years.append(i[1])
years = (set(years))
years
```

```python
# set de meses (sin repetir)
months = []
for i in my_tuple:
    months.append(i[0])
months = (set(months))
months
```

```python
# diccionario indexado en años (listo para llenar)
dict_years = {y: []for y in years}
dict_years
```

```python
# diccionario indexado en meses listo para llenar
dict_months = {m:[] for m in months}
dict_months
```

```python
datos_new['AGO', '2008']
```

```python
# llenar cada año con DICCIONARIO indexado en meses
dict_years_months = {y:dict_months for y in dict_years}
for y in years:
    for m in months:
        dict_years_months[y][m] = datos_new[m,y]
dict_years_months
```

```python
# llenar cada mes con diccionario indexado en años
dict_months_years = {m:dict_years for m in dict_months}
for m in months:
    for y in years:
        dict_months_years[m][y] = datos_new[m,y]
dict_months_years
```
